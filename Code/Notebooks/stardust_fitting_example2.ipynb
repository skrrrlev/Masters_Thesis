```python
import stardust
import numpy as np
from os import chdir

root = '/Volumes/Git/Master_Thesis/'
chdir(root)
```

```python
kwargs = {
        'custom_umin_indices': None,
        'custom_qpah_indices': None,
        'custom_gamma': None,
    }
ctf = stardust.main.ctf(config_file='Data/COSMOS/stardust/cosmos.config', **kwargs)
```

```python
ctf.config
```

```python
idx = 0
ctf.fit_object(idx, False)
ctf.show_fit(idx,radio=True)
```

```python
n_obj = len(ctf.fnu[:,0])
print(f'number of objects: {n_obj}')
n_proc = -1 #Number of threads to use, -1 for all available threads, 1 for serial mode
```

```python
ctf.fit_catalogue(n_proc=n_proc,n_obj=n_obj)
```

```python
idx = 2
ctf.show_fit(idx)

#All the physical parameters like mstar, mdust and lir are contained within the ctf class
#These can be accessed like so
dust_mass = np.log10(ctf.mdust[idx])
e_dust_mass = ctf.e_mdust[idx]/ctf.mdust[idx]*np.log(10)**-1
sfr = np.log10(ctf.tab['sfr_opt'][idx])

print(r'SFR: {:.2f}'.format(sfr))
```

```python
ctf.tab
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,5),dpi=100)

mstar = np.log10(ctf.tab['mstar'])
sfr = np.log10(ctf.tab['sfr_opt'])
z = ctf.tab['z']
ID = ctf.tab['id']
plt.scatter(ctf.tab['mstar'],ctf.tab['sfr_opt'],c=ctf.tab['z'],s=20,vmin=0,vmax=4,cmap='Spectral_r')
plt.colorbar(label='Redshift')
plt.loglog()
plt.xlim(1e8,)
plt.ylim(1e-2,)
plt.ylabel(r'SFR$_{opt}$ [M$_\odot$ yr$^{-1}$]')
plt.xlabel(r'$M_*$[M$_\odot$]')
plt.grid(alpha=.2)
plt.show()

for i in range(n_obj):
    print(f'(ID {ID[i]}) Mstar = {mstar[i]}, optical sfr = {sfr[i]}, z = {z[i]}')
```

```python
from MTLib.cosmos import Catalogue
cc = Catalogue()
M_cosmos = [float(M) for M in cc.get_stellar_masses(ID)]
ez_mass = [np.log10(float(M)) for M in cc.get_property(ID,'ez_mass')]
for i in range(n_obj):
    print(f'(ID {ID[i]}) Fitted Mstar / COSMOS Mstar = {mstar[i]/M_cosmos[i]}')
```

```python
plt.figure(figsize=(10,5),dpi=100)
minn = min(min(mstar),min(M_cosmos))*0.95
maxx = max(max(mstar),max(M_cosmos))*1.05
plt.plot([minn,maxx],[minn,maxx],'k--')
plt.scatter(mstar,M_cosmos,s=20,c='k')
plt.ylabel(r'lp best $M_*$[log$_{10}$M$_\odot$]')
plt.xlabel(r'Fitted $M_*$[log$_{10}$M$_\odot$]')
plt.xlim([minn,maxx])
plt.ylim([minn,maxx])
plt.grid(alpha=.2)
plt.show()
```

```python
plt.figure(figsize=(10,5),dpi=100)
minn = min(min(mstar),min(ez_mass))*0.95
maxx = max(max(mstar),max(ez_mass))*1.05
plt.plot([minn,maxx],[minn,maxx],'k--')
plt.scatter(mstar,ez_mass,s=20,c='k')
plt.ylabel(r'ez 500 $M_*$[log$_{10}$M$_\odot$]')
plt.xlabel(r'Fitted $M_*$[log$_{10}$M$_\odot$]')
plt.xlim([minn,maxx])
plt.ylim([minn,maxx])
plt.grid(alpha=.2)
plt.show()
```
